## Imports and dataset loading

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
import math
import pandas as pd

from sklearn.multioutput import MultiOutputRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_validate

ampNumber   = 3
layerArray  = [2, 3, 4, 5, 6]
neuronArray = [32, 64, 128, 256, 512, 1024]
markers     = ["o", "^", "P", "s", "D", "x"]

trainGainArray = []
trainLossArray = []
testGainArray  = []
testLossArray  = []
binaryDataSet  = []

originalDataSet = pd.read_csv('../datasets/4_channel_type/dataset-{}.txt'.format(ampNumber), sep = ',',header = None)

for line in originalDataSet.values:
    myList = [1 if j != 0 else 0 for j in line[:40]]
    myList.extend(line[40:])
    binaryDataSet.append(myList)

binaryDataSet = pd.DataFrame(binaryDataSet)
gainScaler    = MinMaxScaler(feature_range = (-1, 1))
lossScaler    = MinMaxScaler(feature_range = (-1, 1))

gainScalerAux = []
lossScalerAux = []
    
for j in range(0, ampNumber * 2, 2):
    gainScalerAux.extend(binaryDataSet.values[:, 40 + j])
    lossScalerAux.extend(binaryDataSet.values[:, 40 + j + 1])

gainScaler.fit(np.array(gainScalerAux).reshape(-1, 1))
lossScaler.fit(np.array(lossScalerAux).reshape(-1, 1))

def loadDataset(): 
    dataSet = binaryDataSet.values[:, :40]
    
    for i in range(0, ampNumber * 2, 2):
        gain    = np.array(binaryDataSet.values[:, 40 + i])
        gain    = gainScaler.transform(gain.reshape(-1, 1))
        dataSet = np.hstack((dataSet, gain))
        
        loss    = np.array(binaryDataSet.values[:, 40 + i + 1])
        loss    = lossScaler.transform(loss.reshape(-1, 1))
        dataSet = np.hstack((dataSet, loss))
            
    features, result = np.array(dataSet[:, :40]), np.array(dataSet[:, 40:])
    return features, result

def invertGainNorm(value):
    auxArray = np.array([value, 0, 0, 0, 0, 0]).reshape(-1, 1)
    return gainScaler.inverse_transform(auxArray)[0][0]

def invertLossNorm(value):
    auxArray = np.array([value, 0, 0, 0, 0, 0]).reshape(-1, 1)
    return lossScaler.inverse_transform(auxArray)[0][0]
    
def getGainError(value1, value2):
    return abs(invertGainNorm(value1) - invertGainNorm(value2))

def getLossError(value1, value2):
    return abs(invertLossNorm(value1) - invertLossNorm(value2))
```

## Train and validation functions

```python
from sklearn.metrics import mean_absolute_error

def getTrainErrors(regressor, X, y):
    N_EPOCHS        = 150
    N_BATCH         = 300
    epoch           = 0
    trainGainErrors = []
    trainLossErrors = []
    
    testing = 0
    
    while epoch < N_EPOCHS:
        random_perm      = np.random.permutation(X.shape[0])
        mini_batch_index = 0
        
        while True:
            indices = random_perm[mini_batch_index : mini_batch_index + N_BATCH]
            regressor.partial_fit(X[indices], y[indices])
            mini_batch_index += N_BATCH
            
            if mini_batch_index >= len(X):             
                gainError  = 0
                lossError  = 0
                prediction = regressor.predict(X)

                for predicted, expected in zip(prediction, y):
                    for i in range(0, ampNumber * 2, 2):
                        gainError += getGainError(predicted[i], expected[i]) 
                        lossError += getLossError(predicted[i + 1], expected[i + 1])
                        #print(getGainError(predicted[i], expected[i]))

                trainGainErrors.append((gainError / ampNumber) / len(X))
                trainLossErrors.append((lossError / ampNumber) / len(X))
                break
        
        epoch += 1
        #print(trainGainErrors[-1])
    
    return np.array(trainGainErrors), np.array(trainLossErrors)

def getTestErrors(regressor, X, y, folds = 5):
    foldSize       = math.ceil(X.shape[0] / folds)
    testGainErrors = []
    testLossErrors = []
    
    for i in range(folds): 
        sliceBegin = i * foldSize
        sliceEnd   = (i + 1) * foldSize
        
        X_train = np.delete(X, np.s_[sliceBegin: sliceEnd], 0)
        y_train = np.delete(y, np.s_[sliceBegin: sliceEnd], 0)
        X_test  = X[sliceBegin: sliceEnd]
        y_test  = y[sliceBegin: sliceEnd]

        regressor.fit(X_train, y_train)
        
        gainError  = 0
        lossError  = 0
        prediction = regressor.predict(X_test)
        
        for predicted, expected in zip(prediction, y_test):
            for i in range(0, ampNumber * 2, 2):
                gainError += getGainError(predicted[i], expected[i]) 
                lossError += getLossError(predicted[i + 1], expected[i + 1])
                 
        testGainErrors.append((gainError / ampNumber) / foldSize)
        testLossErrors.append((lossError / ampNumber) / foldSize)
        
    return np.array(testGainErrors), np.array(testLossErrors)
```

```python
def trainModel(strategy):    
    features, result       = loadDataset()
    gainErrors, lossErrors = getTrainErrors(regressor, features, result)
    
    print("Train error")
    print(gainErrors, "=> %0.2f (+/- %0.2f)" % (np.mean(gainErrors), gainErrors.std() * 2))
    print(lossErrors, "=> %0.2f (+/- %0.2f)" % (np.mean(lossErrors), lossErrors.std() * 2))
    print()
    
    return np.mean(gainErrors), np.mean(lossErrors)

def testModel(strategy):    
    features, result       = loadDataset()
    gainErrors, lossErrors = getTestErrors(strategy, features, result)
    
    print("Test errors")
    print(gainErrors, "=> %0.2f (+/- %0.2f)" % (np.mean(gainErrors), gainErrors.std() * 2))
    print(lossErrors, "=> %0.2f (+/- %0.2f)" % (np.mean(lossErrors), lossErrors.std() * 2))
    print()
    
    return np.mean(gainErrors), np.mean(lossErrors)

def setMLP(layers, neuronNumber):
    if layers == 2:
        return MLPRegressor(hidden_layer_sizes = (neuronNumber, neuronNumber), activation = 'identity', learning_rate = 'constant', alpha = 0.001, random_state = 0)
    
    if layers == 3:
        return MLPRegressor(hidden_layer_sizes = (neuronNumber, neuronNumber, neuronNumber), activation = 'identity', learning_rate = 'constant', alpha = 0.001, random_state = 0)
    
    if layers == 4:
        return MLPRegressor(hidden_layer_sizes = (neuronNumber, neuronNumber, neuronNumber, neuronNumber), activation = 'identity', learning_rate = 'constant', alpha = 0.001, random_state = 0)
    
    if layers == 5:
        return MLPRegressor(hidden_layer_sizes = (neuronNumber, neuronNumber, neuronNumber, neuronNumber, neuronNumber), activation = 'identity', learning_rate = 'constant', alpha = 0.001, random_state = 0)
    
    if layers == 6:
        return MLPRegressor(hidden_layer_sizes = (neuronNumber, neuronNumber, neuronNumber, neuronNumber, neuronNumber, neuronNumber), activation = 'identity', learning_rate = 'constant', alpha = 0.001, random_state = 0)
```

## Plotting functions

```python
def plotGainError():
    fig, axis = plt.subplots(figsize = (10, 5))

    for i in range(len(layerArray)):
        axis.plot(neuronArray, trainGainArray[i])
        axis.scatter(neuronArray, trainGainArray[i], marker = markers[i])
    
    axis.set_title("Gain error for {} amplifiers".format(ampNumber))
    axis.set_xlabel("Neuron number")
    axis.set_ylabel("MAE (dB)")
    axis.legend(["{} hidden layers".format(x) for x in layerArray])

    plt.show()

def plotLossError():
    fig, axis = plt.subplots(figsize = (10, 5))

    for i in range(len(layerArray)):
        axis.plot(neuronArray, trainLossArray[i])
        axis.scatter(neuronArray, trainLossArray[i], marker = markers[i])
    
    axis.set_title("Loss error for {} amplifiers".format(ampNumber))
    axis.set_xlabel("Neuron number")
    axis.set_ylabel("MAE (dB)")
    axis.legend(["{} hidden layers".format(x) for x in layerArray])

    plt.show()
```

## Putting all together

```python
for layers in layerArray:
    print(layers, "layers")
    
    trainGainArray.append([])
    trainLossArray.append([])
    
    testGainArray.append([])
    testLossArray.append([])
    
    for neurons in neuronArray:
        regressor = setMLP(layers, neurons)
        
        trainGain, trainLoss = trainModel(regressor)
        trainGainArray[-1].append(trainGain)
        trainLossArray[-1].append(trainLoss)
        
        #testGains, testLosses = testModel(regressor)
        #testGainArray[-1].append(testGains)
        #testLossArray[-1].append(testLosses)
```
